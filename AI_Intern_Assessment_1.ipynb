```python
!pip install chromadb sentence-transformers

import chromadb
from sentence_transformers import SentenceTransformer
import string

documents = [
    "Artificial Intelligence is transforming modern industries.",
    "Machine learning is a subset of artificial intelligence.",
    "Python is widely used for data science and AI development.",
    "ChromaDB is a vector database used for embedding storage.",
    "Retrieval Augmented Generation improves LLM responses."
]

def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    return text

processed_docs = [preprocess(doc) for doc in documents]

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(processed_docs)

client = chromadb.Client()

collection = client.get_or_create_collection(
    name="mini_rag"
)

for i, doc in enumerate(processed_docs):
    collection.add(
        documents=[doc],
        embeddings=[embeddings[i]],
        ids=[str(i)]
    )

query  = input("Enter your query: ")

processed_query = preprocess(query)
query_embedding = model.encode([processed_query])

results = collection.query(
    query_embeddings=query_embedding.tolist(),
    n_results=1
)

retrived_doc = results["documents"][0][0]

print("\n most relavent documents: ")
print(retrived_doc)

```
